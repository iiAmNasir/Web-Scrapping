##Import Libraries

```
import requests
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup
```

##Scrapping all links Of Asia International news

```
#A method that will check if url is valid or not, Return 1 if valid otherwise it return 0
def Is_Valid_Url(url):
    ParseUrl = urlparse(url)
    return bool(ParseUrl.netloc) and bool(ParseUrl.scheme)
```

```

Final_urls = set()

#count visisted urls
count = 0

# A function that will return all the urls of the same website found inside given 'url'
def get_links(url):
    urls = set()
    domain = urlparse(url).netloc
    soup = BeautifulSoup(requests.get(url).content, "html.parser")

    for link in soup.findAll("a"):
        href = link.attrs.get("href")
        #ignore if a tag is empty
        if href == "" or href is None:
            continue

        # join if the URL is relative
        href = urljoin(url, href)
        parsed_href = urlparse(href)

        # remove URL GET parameters, URL fragments, etc.
        href = parsed_href.scheme + "://" + parsed_href.netloc + parsed_href.path
        #IGNORE IF URL IS VALID
        if not Is_Valid_Url(href):
            continue
        # Ignbore if url is already in list   
        if href in Final_urls:
            continue
        #Ignore if url is out of that domain
        if domain not in href:
            continue
        urls.add(href)       
        Final_urls.add(href)
    return urls
```

```
def Scrape(url, max_urls=50):
    global count
    count += 1
    links = get_links(url)
    for link in links:
        if count > max_urls:
            break
        Scrape(link, max_urls)
```

```
def Display():
  k=1
  for link in Final_urls:
    print(k, link)
    k=k+1
```

```

if __name__ == "__main__":
    Scrape("https://aninews.in/",500)
    print(" Total Links Found :",  len(Final_urls))
    Display()
```
